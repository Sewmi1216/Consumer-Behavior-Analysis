```python
import math

features = ['Age', 'Review Rating','Previous Purchases'] + [col for col in df.columns if col.startswith('Frequency of Purchases_')]

# Define subplot layout (3 columns)
num_features = len(features)
num_rows = math.ceil(num_features / 3)  # Calculate required rows

# Create figure with subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 5 * num_rows)) 

# Flatten the axes array for easier iteration
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.regplot(x=df[feature], y=df['Purchase Amount (USD)'], ax=axes[i], line_kws={'color': 'red'})
    axes[i].set_title(f'Regression Plot: {feature} vs Purchase Amount')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()

# Save the figure as a JPEG file
plt.savefig('regression_plots.jpg', format='jpeg', dpi=300)

# Show the plot
plt.show()
```

```python
# Select relevant numerical features for clustering
features = ['Age', 'Purchase Amount (USD)', 'Previous Purchases']
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize numerical features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])
```

```python
from sklearn.cluster import KMeans

wcss = []  # Within-cluster sum of squares
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')

# Save the figure as a JPEG file
plt.savefig('elbow_curve.jpg', format='jpeg', dpi=300)

# Show the plot
plt.show()
```

```python
# Choosing the optimal K (from the elbow plot)
optimal_k = 4

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Cluster summary statistics
cluster_summary = df.groupby('Cluster')[features].mean()
print(cluster_summary)

# Visualizing Clusters (Age vs. Purchase Amount)
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Age'], y=df['Purchase Amount (USD)'], hue=df['Cluster'], palette='viridis', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.title('Customer Segmentation (K-Means Clustering)')
plt.legend(title='Cluster')

# Save the figure as a JPEG file
plt.savefig('cluster.jpg', format='jpeg', dpi=300)

# Show the plot
plt.show()
```
